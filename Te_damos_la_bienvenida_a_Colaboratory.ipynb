```
import random

usuario = input("piedra, papel o tijeras: ")
computadora = random.choice(["piedra", "papel", "tijeras"])

print("PC:", computadora)

if usuario == computadora:
    print("Empate")
elif (usuario == "piedra" and computadora == "tijeras") or (usuario == "papel" and computadora == "piedra") or (usuario == "tijeras" and computadora == "papel"):
    print("Ganaste")
else:
    print("Perdiste")
```
